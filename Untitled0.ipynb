```python
#คํานวณพื้นที่ของที่ดินรูปสี่เหลี่ยม จากความกว้างและความยาว (หน่วยเป็นเมตร) แสดงพื้นที่ในหน่วยตารางเมตร และตารางวา
#User
l = float(input("Lenght in m : "))
w = float(input("Width in m : "))
#Calculate
area = l*w
wa = area/4
#Output
print()
print("Area in m^2 :",area)
print("Area in wa :",wa)
```

```python
#Convert temperature in Celsius to Fahrenheit
#User
c = float(input("Temperature in Celsius : "))
#Calculate
f = (c*9/5)+32
#Output
print()
print("Temperature in Fahrenheit :",f," F")
```

```python
#Calculate circumference and area of a circle from its diameter (not radius)
pi = 3.141
#User
d = float(input("Dimeter in m : "))
#Calculate
circumference = pi*d
area = pi*(d**2)
#Output
print()
print("Circumference : ",circumference," m")
print("Area : ",area," m^2")
```

```python
#Calculate Body Mass Index (BMI) from weight and height
#User
w = float(input("Weight in kg : "))
h = float(input("Height in m : "))
#Calculate
bmi = w/(h**2)
#Output
print()
print("BMI : ",bmi)
```
